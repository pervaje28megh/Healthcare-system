```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
# Read the subscriber Details
df = pd.read_csv('subscriber.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Replace the null values
df["first_name"].fillna("NA", inplace = True)
```

```python
# Replace the null values
df["Phone"].fillna("NA", inplace = True)
df["Subgrp_id"].fillna("NA", inplace = True)
df["Elig_ind"].fillna("N", inplace = True)
df.head(60)
```

```python
df.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
```

```python
# Check Subscriber Id not less than 9 digit
count = 0
for i in df['sub_id'].values:
    
    if len(i)<9:
        df.drop([count], axis=0, inplace=True)
    elif len(str(i))>10:
        df.drop([count], axis=0, inplace=True)
    count = count+1
```

```python
df['Elig_ind'].unique()
```

```python
# Check always eff_date is greater than term_date 
count = 0
for x,y in df[['eff_date','term_date']].values:
    dob1 = datetime.strptime(x,'%d-%m-%Y').date()
    dob2 = datetime.strptime(y,'%d-%m-%Y').date()
    if dob1 > dob2:
        df.drop([count], axis=0, inplace=True)
    count = count + 1
```

```python
# Export the Data
df.to_csv('subscriber_Cleaned.csv',index=False)
```
